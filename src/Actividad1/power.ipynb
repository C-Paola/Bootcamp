## Descripción del dataset

El Global Power Plant Database es un conjunto de datos abierto y completo que recopila información sobre plantas de generación de energía eléctrica en todo el mundo. Publicado por el World Resources Institute (WRI) y alojado en Kaggle (Global Power Plant Database), este dataset contiene 34936 filas y 36 columnas, representando plantas de energía de diversos tipos (carbón, gas, hidroeléctrica, solar, nuclear, entre otros) en múltiples países. Cada fila corresponde a una planta de energía, con atributos que describen su ubicación geográfica, capacidad de generación, tipo de combustible y otras características operativas. Este dataset es especialmente valioso para análisis energéticos, ambientales y espaciales, ya que incluye coordenadas de latitud y longitud. El dataset es ideal para explorar patrones globales en la producción de energía, evaluar la transición hacia fuentes renovables y analizar la distribución geográfica de infraestructura energética.

## Posibles análisis
1. Distribución Geográfica de Plantas de Energía por Tipo de Combustible
   
   Variables potencialmente interesantes:
   1. **latitude y longitude:** Coordenadas geográficas para ubicar cada planta en un mapa, permitiendo visualizaciones como mapas de calor o dispersión. 
   2. **primary_fuel:** Tipo de combustible principal, clave para clasificar y colorear las plantas en el mapa según su fuente energética. 
   3. **country:** Permite agregar un nivel de análisis por país, identificando líderes en energías renovables o dependientes de combustibles fósiles.
2. Predicción de la Capacidad de Generación de Energía
   
   Variables potencialmente interesantes:
   1. **capacity_mw:** Variable objetivo numérica que representa la capacidad instalada de la planta, a predecir. 
   2. **primary_fuel:** Tipo de combustible, un predictor clave ya que la tecnología influye en la capacidad (e.g., nuclear vs. solar). 
   3. **latitude y longitude:** Pueden capturar efectos geográficos, como regiones con mayor potencial solar o hidroeléctrico. 
3. Segmentación de Plantas por Características Operativas y Geográficas
   
   Variables potencialmente interesantes:
   1. **capacity_mw:** Define el tamaño de la planta, útil para diferenciar entre pequeñas y grandes instalaciones. 
   2. **primary_fuel:** Permite segmentar por tecnología energética (renovable vs. no renovable). 
   3. **latitude y longitude:** Incorporan la dimensión espacial, agrupando plantas cercanas o en regiones similares.
4. Análisis de Tendencias Energéticas por Región
   
   Variables potencialmente interesantes:
   1. **country:** Agrupa plantas por país o región, base del análisis comparativo. 
   2. **primary_fuel:** Muestra la mezcla energética por región (e.g., porcentaje de solar vs. carbón). 
   3. **capacity_mw:** Evalúa la escala total de producción por tipo de combustible en cada región.
