## Descripción del dataset
El **Bike Sharing Demand** es un conjunto de datos creado para una competencia en Kaggle basado en datos reales del sistema de alquiler de bicicletas Capital Bikeshare en Washington D.C. entre 2011 y 2012. El dataset contiene está dividido en dos partes: el conjunto de entrenamiento (train.csv) con 10,886 filas y 15 variables, y el conjunto de prueba (test.csv) con 6,493 filas y 9 columnas. El objetivo principal es predecir la demanda de bicicletas por hora (count) en función de factores como el clima, la fecha y el tipo de día. 

Cada fila representa una hora específica, con atributos que describen las condiciones temporales (e.g., fecha, hora), climáticas (e.g., temperatura, clima) y la cantidad de alquileres (count, dividido en usuarios casuales y registrados). Este dataset es ideal para explorar patrones de movilidad urbana, predecir demanda y analizar influencias ambientales, ofreciendo oportunidades para aplicar técnicas de regresión, series temporales y visualización en un contexto comercial y analítico relevante.

#### Posibles análisis

1. Predicción de la Demanda de Bicicletas por Hora  
   
   Variables potencialmente interesantes: 
     - **count**: Variable objetivo numérica, suma de alquileres casuales y registrados, a predecir.  
     - **datetime (y derivados: hour, weekday)**: Hora y día de la semana, capturan patrones horarios y semanales de uso.  
     - **temp**: Temperatura en grados Celsius, afecta la disposición a usar bicicletas.  
     - **weather**: Condición climática (e.g., Clear, Rain), un predictor clave de la demanda.  
     - **workingday**: Indica si es día laboral (1) o no (0), influye en patrones de uso.  

2. Análisis de Patrones de Uso por Condiciones Climáticas  
   
   Variables potencialmente interesantes:
     - **count**: Número de alquileres, base para comparación según clima.  
     - **weather**: Tipo de clima, permite segmentar demanda por condiciones (e.g., lluvia vs. soleado).  
     - **temp**: Temperatura, mide efectos térmicos en el uso.  
     - **humidity**: Humedad relativa, puede influir en la comodidad de los usuarios.  

3. Segmentación de Horas por Patrones de Demanda  
   
   Variables potencialmente interesantes:
     - **count**: Demanda total, define el tamaño del uso por hora.  
     - **hour**: Hora del día, captura picos diarios (e.g., mañana vs. noche).  
     - **weekday**: Día de la semana, diferencia entre días laborales y fines de semana.  
     - **holiday**: Indica feriados, útil para segmentar días especiales.  

4. Análisis de Tendencias Temporales en la Demanda de Bicicletas
   
   Variables potencialmente interesantes:
     - **count**: Demanda de alquileres, base para analizar tendencias.  
     - **year**: Año (2011 o 2012), muestra cambios interanuales.  
     - **month**: Mes, captura estacionalidad (e.g., verano vs. invierno).  
     - **hour**: Hora, permite análisis intradía dentro de tendencias más amplias.
