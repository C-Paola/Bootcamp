## Descripcin del dataset

Recopila informacin detallada sobre ventas de casas residenciales en Ames, Iowa, entre 2006 y 2010. El dataset 
contiene 2919 filas y 80 variables originalmente viene dividido en instancias de training y testing. Cabe aclarar que el subconjunto para testing no incluye la variable de "SalePrice" ya que este dataset es una competencia actual en Kaggle.

Cada fila representa una casa vendida, con atributos que describen caractersticas fsicas (e.g., tamao, nmero de habitaciones), calidad (e.g., materiales, condicin), ubicacin (e.g., vecindario) y el precio de venta (SalePrice). Este dataset es ideal para explorar factores que influyen en los precios inmobiliarios, desarrollar modelos predictivos y analizar patrones en el mercado de viviendas, ofreciendo oportunidades para aplicar tcnicas de regresin, visualizacin y segmentacin en un contexto comercial y analtico relevante.

## Posibles Anlisis y Atributos Relevantes

4. Anlisis de Tendencias Temporales en Precios de Vivienda
